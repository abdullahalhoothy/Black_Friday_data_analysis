```python
#!pip install pandas
import numpy as np
import pandas as pd
df = pd.read_csv(r'C:\Users\abdullah-laptop\Downloads\Black_driday_data_analysis-master\Black_driday_data_analysis-master\Data Test\train.csv')
import matplotlib.pyplot as plt 


for col_name in df.columns:
    print(col_name, len(df[col_name].unique()))
```

```python
# how many "Entries" do we have based on each column 

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=[15, 15])

df['Gender'].value_counts().plot(kind='barh', ax=axes[0,0], title='Gender')
df['Age'].value_counts().plot(kind='barh', ax=axes[0,1], title='Age')
df['City_Category'].value_counts().plot(kind='barh', ax=axes[1,0], title='City_Category')
df['Marital_Status'].value_counts().plot(kind='barh', ax=axes[1,1], title='Marital_Status')
df['Occupation'].value_counts().plot(kind='barh', ax=axes[2,0], title='Occupation')
df['Stay_In_Current_City_Years'].value_counts().plot(kind='barh', ax=axes[2,1], title='Stay_In_Current_City_Years')
```

```python
# how many "unique Users" do we have based on each column 
df4= df.drop_duplicates('User_ID')

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=[15, 15])

df4['Gender'].value_counts().plot(kind='barh', ax=axes[0,0], title='Gender')
df4['Age'].value_counts().plot(kind='barh', ax=axes[0,1], title='Age')
df4['City_Category'].value_counts().plot(kind='barh', ax=axes[1,0], title='City_Category')
df4['Marital_Status'].value_counts().plot(kind='barh', ax=axes[1,1], title='Marital_Status')
df4['Occupation'].value_counts().plot(kind='barh', ax=axes[2,0], title='Occupation')
df4['Stay_In_Current_City_Years'].value_counts().plot(kind='barh', ax=axes[2,1], title='Stay_In_Current_City_Years')
```

```python
# how much has each unique group(of each column) purchased 
#first identify all possible groups 
for col_name in ['Gender', 'Age', 'Occupation', 'City_Category','Stay_In_Current_City_Years','Marital_Status']:
    print(sorted(df[col_name].unique()))
    
# Second for each unique group, we do a group by to sum the purchase amount
dfGender = df.groupby(['Gender'])['Purchase'].sum()
dfage = df.groupby(['Age'])['Purchase'].sum()
dfoccu = df.groupby(['Occupation'])['Purchase'].sum()
dfCC = df.groupby(['City_Category'])['Purchase'].sum()
dfstay = df.groupby(['Stay_In_Current_City_Years'])['Purchase'].sum()
dfM = df.groupby(['Marital_Status'])['Purchase'].sum()

dfGender.sort(['A', 'B'], ascending=[1, 0])

# Now we plot
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=[15, 15])
dfGender.plot(kind='barh', ax=axes[0,0], title='Gender')
dfage.plot(kind='barh', ax=axes[0,1], title='Age')
dfCC.plot(kind='barh', ax=axes[1,0], title='City_Category')
dfM.plot(kind='barh', ax=axes[1,1], title='Marital_Status')
dfoccu.plot(kind='barh', ax=axes[2,0], title='Occupation')
dfstay.plot(kind='barh', ax=axes[2,1], title='Stay_In_Current_City_Years')
```

```python
# lets compare the counts of entiers and unique users and the amount of purchases 
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=[15, 15])

# count of eneries 
df['Gender'].value_counts().plot(kind='barh', ax=axes[0,0], title='Gender')
df['Age'].value_counts().plot(kind='barh', ax=axes[1,0], title='Age')
df['City_Category'].value_counts().plot(kind='barh', ax=axes[2,0], title='City_Category')
df['Marital_Status'].value_counts().plot(kind='barh', ax=axes[3,0], title='Marital_Status')
df['Occupation'].value_counts().plot(kind='barh', ax=axes[4,0], title='Occupation')
df['Stay_In_Current_City_Years'].value_counts().plot(kind='barh', ax=axes[5,0], title='Stay_In_Current_City_Years')
# unique user count
df4['Gender'].value_counts().plot(kind='barh', ax=axes[0,1], title='Gender')
df4['Age'].value_counts().plot(kind='barh', ax=axes[1,1], title='Age')
df4['City_Category'].value_counts().plot(kind='barh', ax=axes[2,1], title='City_Category')
df4['Marital_Status'].value_counts().plot(kind='barh', ax=axes[3,1], title='Marital_Status')
df4['Occupation'].value_counts().plot(kind='barh', ax=axes[4,1], title='Occupation')
df4['Stay_In_Current_City_Years'].value_counts().plot(kind='barh', ax=axes[5,1], title='Stay_In_Current_City_Years')
# amount of purchaes 
dfGender.sort_values(ascending=False).plot(kind='barh', ax=axes[0,2], title='Gender Purchase Billions$')
dfage.sort_values(ascending=False).plot(kind='barh', ax=axes[1,2], title='Age Purchase Billions$')
dfCC.sort_values(ascending=False).plot(kind='barh', ax=axes[2,2], title='City_Category Purchase Billions$')
dfM.sort_values(ascending=False).plot(kind='barh', ax=axes[3,2], title='Marital_Status Purchase Billions$')
dfoccu.sort_values(ascending=False).plot(kind='barh', ax=axes[4,2], title='Occupation Purchase Billions$')
dfstay.sort_values(ascending=False).plot(kind='barh', ax=axes[5,2], title='Stay_In_Current_City_Years Purchase Billions$')
```
